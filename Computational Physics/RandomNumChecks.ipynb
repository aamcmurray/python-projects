```python
import numpy as np 
import matplotlib.pyplot as plt
np.random.seed(100)

# linear congruent method of random number generation
def random_num_gen(c,a,M,r):
    r=(a*r+c)%M
    return r
# Using bad values
c=57
a=1
M=256
r=10

r_values=[]
x=np.linspace(0,500,500)

for i in range(0,500):
    r=random_num_gen(c,a,M,r)
    r_values.append(r)

plt.plot(x,r_values,'.', label='linear congruent method random numbers')
plt.xlabel('sequence number, i')
plt.ylabel('random number, r')
plt.legend()
```

```python
# tortoise-hare algorithm for finding where a sequence repeats

def findDuplicate(nums):
    tortoise=nums[0]
    hare=nums[0]
    while True:
        tortoise=nums[tortoise]
        hare=nums[nums[hare]]
        if tortoise == hare:
            break
    ptr1=nums[0]
    ptr2=tortoise
    while ptr1!=ptr2:
        ptr1=nums[ptr1]
        ptr2=nums[ptr2]
    return ptr1
```

```python
findDuplicate(r_values) # duplicate value 67 is the first repitition
```

### further tests for uniformity of distribution of random numbers

#### evaluate the kth moment of a distribution

$<x^k>=\frac{1}{N} \sum \limits_{i}^{N} x^k_i $

If the numbers are distributed uniformly then the kth moment of the distribution is approximately the moment of the distribution function P(x).

$1/(k+1)+O(1/sqrt(N))$

```python
c=int('13',8) # 13(base 8)
a=int('2736731673163155',8)
M=2**48
r=1
N=100
scale=10e14
r_values=[]
x=np.linspace(0,N,N)

for i in range(0,N):
    r=random_num_gen(c,a,M,r)
    r_values.append(r/scale)

plt.plot(x,r_values,'.', label='linear congruent method random numbers')
plt.xlabel('sequence number, i')
plt.ylabel('random number, r')
plt.legend()

def powerk(list,kay):
    return [i ** kay for i in list]
k=[1,3,7]

# check for order 1

for i in range(0,len(k)):
    print(np.sqrt(N)*np.abs((1/N)*(np.sum(powerk(r_values,k[i])))-(1/(k[i]+1))))
    
```
