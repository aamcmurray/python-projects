# Sum of a series

```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
```

```python
def summation_machine(a,b,c):
	dictionary_one={}
	convergence=list()
	x=np.linspace(a,b,1000)
	for j in range(0,1000):
		nth_term=x[j]
		summation=x[j]
		i=2
		while abs(nth_term/summation) > c:
			nth_term=nth_term*(-x[j]**2)/(((2*i)-1)*((2*i)-2))
			summation=summation+nth_term
			imax=i
			convergence.append(abs(nth_term/summation))
			i+=1
		dictionary_one[x[j]]=str(summation)
	return dictionary_one,convergence

def sin_function_method(a,b):
	dictionary_two={}
	x=np.linspace(a,b,1000)
	for j in range(0,1000):
		true_value=np.sin(x[j])
		dictionary_two[x[j]]=str(true_value)
	return dictionary_two

def dictionary_convert(adict):
	ind=list()
	val=list()
	for key in adict:
		ind.append(key)
		val.append(float(adict[key]))
	return ind, val
```

```python
a=np.deg2rad(0.01)
b=np.deg2rad(360)
c=10e-8
first_dict,conv=(summation_machine(a,b,c))

if conv[len(conv)-1] < c:
    print(True)

indices,values=dictionary_convert(first_dict)
indices_two,values_two=dictionary_convert((sin_function_method(a,b)))

plt.plot(indices,values, label='Summation')
plt.plot(indices_two,values_two, '--', label='True value')
plt.legend()
plt.show()
```

The summation converges to within the tolerance and produces the correct answer.
