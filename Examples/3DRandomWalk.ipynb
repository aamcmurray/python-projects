```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def normalised_steps():
    dx = 2*(random.random()-0.5)
    dy = 2*(random.random()-0.5)
    dz = 2*(random.random()-0.5)
    L = np.sqrt(dx**2+dy**2+dz**2)
    dx = dx/L
    dy = dy/L
    dz = dz/L
    return dx, dy, dz

def walk(n):
    x = 0
    y = 0
    z = 0
    for i in range(n):
        dx, dy, dz = normalised_steps()
        x += dx
        y += dy
        z += dz
        yield x,y,z
    return

def plotting(values):
    x = []
    y = []
    z = []
    for i in range(0, len(values)):
        x.append(values[i][0])
        y.append(values[i][1])
        z.append(values[i][2])
    return x,y,z

steps=1000
walker = list(walk(steps))
x,y,z = plotting(walker)

fig = plt.figure(figsize=(10,10),dpi=200)
ax = fig.add_subplot(111, projection='3d')
ax.grid(True)
ax.xaxis.pane.fill = ax.yaxis.pane.fill = ax.zaxis.pane.fill = True
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.scatter3D(x,y,z, c='red', alpha=0.25,s=2)
ax.plot3D(x,y,z, c='blue', alpha=0.5, lw=0.5)
plt.title('3D Random Walk')
```
